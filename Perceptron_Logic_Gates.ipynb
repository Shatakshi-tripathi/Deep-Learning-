```python
input=[(0,0),(0,1),(1,0),(1,1)]
```

```python
#AND IMPLEMENTATION
for a,b in input:
    s=a*1+b*1
    if s>=1.5:
        y=1
    else:
        y=0
    print(a,b," ",y)
```

```python
#OR IMPLENTATION
for a,b in input:
    s=a*1+b*1
    if s>=0.5:
        y=1
    else:
        y=0
    print(a,b," ",y)
```

```python
#NAND IMPLENTATION
for a,b in input:
    s=a*(-1)+b*(-1)
    if s>=-1.5:
        y=1
    else:
        y=0
    print(a,b," ",y)
```

```python
#NOR IMPLENTATION
for a,b in input:
    s=a*(-1)+b*(-1)
    if s>=-0.5:
        y=1
    else:
        y=0
    print(a,b," ",y)
```
